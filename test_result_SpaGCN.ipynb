```python
from train_SpaGCN import train_spg
from st_datasets.dataset import *
from tools import evaluate

result = []
result_refine = []
```

# human breast cancer

```python
adata, n_cluster = get_data(dataset_func=get_human_breast_cancer_data)
train_spg(adata, n_cluster)
ari, ari_refine = evaluate(adata, pred_list=['pred', 'refined_pred'])
adata.obs['pred'].to_csv('result/predict/SpaGCN/human_breast_cancer.txt', index=True, sep='\t')
result.append(ari)
result_refine.append(ari_refine)
```

# DLPFC

```python
for i in range(12):
    adata, n_cluster = get_data(dataset_func=get_dlpfc_data, id=i)
    train_spg(adata, n_cluster)
    ari, ari_refine = evaluate(adata, pred_list=['pred', 'refined_pred'])
    adata.obs['pred'].to_csv(f'result/predict/SpaGCN/dlpfc{i}.txt', index=True, sep='\t')
    result.append(ari)
    result_refine.append(ari_refine)
```
